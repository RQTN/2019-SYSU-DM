## Exercise1

```python
import numpy as np
import pandas as pd
import math

# throw how much(N) points
N_array = np.array([20, 50, 100, 200, 300, 500, 1000, 5000])
# repeat times
times = 20        

def estimate_pi(N):
    est_pi = []
    for i in range(times):
        x, y = np.random.random(N), np.random.random(N)
        dist = pow(x**2+y**2, 0.5)
        in_circle = (dist <= 1.0).sum()
        est_pi.append(4*in_circle/N)
    return [np.mean(est_pi), np.var(est_pi)]   

datas = []
for i in range(N_array.size):
    datas.append(estimate_pi(N_array[i]))
    
rows = [('N=%d' % (i)) for i in N_array]
pd.DataFrame(datas, columns=('mean', 'var'), index=rows)
```

## Exercise 2

`x` 通过均匀分布采样获得

```python
import numpy as np
import pandas as pd

# sample how much(N) x
N_array = np.array([5, 10, 20, 30, 40, 50, 60, 70, 80, 100])
# repeat times
times = 20

def estimate_single_integral(N):
    est_s_i = []
    for i in range(times):
        x = np.random.random(N)
        y = x**3
        area = y*(1-0)
        est_s_i.append(area.sum()/N)
    return [np.mean(est_s_i), np.var(est_s_i)] 

datas = []
for i in range(N_array.size):
    datas.append(estimate_single_integral(N_array[i]))
rows = [('N=%d' % (i)) for i in N_array]
pd.DataFrame(datas, columns=('mean', 'var'), index=rows)
```

## Exercise 3

`x` 和 `y` 依然采用均匀分布采样

```python
import numpy as np
import pandas as pd

# sample how much(N) points (x,y)
N_array = np.array([20, 50, 100, 200, 300, 500, 1000, 5000])
# repeat times
times = 100

def estimate_double_integral(N):
    est_d_i = []
    for i in range(times):
        x, y = np.random.uniform(2, 4, N), np.random.uniform(-1, 1, N)
        z = ((y**2)*np.exp(-y**2)+(x**4)*np.exp(-x**2))/(x*np.exp(-x**2))
        volume = z*(4-2)*(1-(-1))
        est_d_i.append(volume.sum()/N)
    return [np.mean(est_d_i), np.var(est_d_i)]

datas = []
for i in range(N_array.size):
    datas.append(estimate_double_integral(N_array[i]))
import pandas as pd
rows = [('N=%d' % (i)) for i in N_array]
pd.DataFrame(datas, columns=('mean', 'var'), index=rows)
```
